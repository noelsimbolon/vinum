# Distribution Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/wine.csv')
```

```python
def visualize_column(column_name: str) -> None:
    """
    Visualize the column in the form of a histogram (with kernel density estimation) and a box plot.
    :param column_name: The column name in the dataframe that is to be visualized.
    :return: None
    """
    # Create a figure with subplots for the histogram and box plot
    fig, axs = plt.subplots(nrows=2,
                            figsize=(5, 7))

    # Add space between subplots
    fig.subplots_adjust(hspace=0.5)

    # Create the histogram on the left subplot
    sns.histplot(data=df,
                 x=column_name,
                 kde=True,
                 ax=axs[0])

    # Create the box plot on the right subplot
    sns.boxplot(data=df,
                x=column_name,
                width=0.2,
                fliersize=3,
                linewidth=1,
                ax=axs[1],
                flierprops={"marker":"o", "markerfacecolor":"none"})

    # Customize the styling of the histogram
    axs[0].set_xlabel(column_name.title())
    axs[0].set_ylabel("Frequency")
    axs[0].set_title("Histogram with Kernel Density Estimation")

    # Customize the styling of the box plot
    axs[1].set_xlabel(column_name.title())
    axs[1].set_title("Box Plot")

    # Show the figure
    plt.show()
```

<p></p>

## Fixed Acidity

```python
visualize_column("fixed acidity")
```

Key points that can be inferred from the histogram and box plot:

1. The majority of the samples have a value of fixed acidity around $7$
2. The median score is slightly above $7$, close to the peak of the histogram
3. The majority of the samples have a value of fixed acidity within a relatively narrow range
4. There are a few outliers beyond the upper whisker and below the lower whisker
