```python
import pandas as pd

df = pd.read_csv('../data/wine.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Fixed Acidity

```python
# Create a figure with subplots for the histograms and box plots
fig, axs = plt.subplots(ncols=2,
                        figsize=(15, 5))

# Add space between subplots
fig.subplots_adjust(wspace=0.3)

# Create the histograms on the left subplot
sns.histplot(data=df,
             x="fixed acidity",
             kde=True,
             ax=axs[0])

# Create the box plots on the right subplot
sns.boxplot(data=df,
            y="fixed acidity",
            width=0.1,
            fliersize=3,
            linewidth=1,
            ax=axs[1])

# Customize the styling of the histograms
axs[0].set_xlabel("Fixed Acidity")
axs[0].set_ylabel("Frequency")
axs[0].set_title("Histogram with Kernel Density Estimation")

# Customize the styling of the box plots
axs[1].set_ylabel("Fixed Acidity")
axs[1].set_title("Box Plot")

# Show the figure
plt.show()
```
